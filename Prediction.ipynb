convert the boolean into 1 or 0

```python
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()
for i in objects_col:
    data[i] = LE.fit_transform(data[i])
    if i == 'Stage':
        continue
data
```

adjustment Stage values

```python
#Define
Stage_values = {'Stage': {2: 3, 0: 1, 3:4, 1:2}}

#replace
data.replace(Stage_values, inplace=True)
data
```

Visualize

```python
import matplotlib.pyplot as plt
import seaborn as sns

exclude_columns = ['Column_Name_1', 'Column_Name_2']  # Declare blank column

columns_to_plot = [col for col in data.columns if col not in exclude_columns]

num_plots = len(columns_to_plot)
ncols = 3
nrows = (num_plots + ncols - 1) // ncols 

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(9, 30))
axes = axes.flat

color_cycle = list(plt.rcParams['axes.prop_cycle'])
num_colors = len(color_cycle)

for i, num_col in enumerate(columns_to_plot):
    sns.histplot(data,
                 x=num_col,
                 stat='count',
                 kde=True,
                 color=color_cycle[i % num_colors]["color"],
                 line_kws={'linewidth': 2, 'linestyle': 'dashed'},
                 alpha=0.4,
                 ax=axes[i])
    sns.rugplot(data,
                x=num_col,
                color=color_cycle[i % num_colors]["color"],
                ax=axes[i], alpha=0.7)
    axes[i].set_xlabel(" ")
    axes[i].set_ylabel("Count", fontsize=7, fontweight='bold', color='black')
    axes[i].set_title(num_col, fontsize=8, fontweight='bold', color='black')
    axes[i].tick_params(labelsize=6)

# Hide empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

fig.suptitle('Distribution of Data', fontsize=12, fontweight='bold', color='darkred', y=0.92)
fig.tight_layout()
fig.subplots_adjust(top=0.9)
plt.show()
```
